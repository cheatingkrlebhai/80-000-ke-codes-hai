```python
clean.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
```

```python
for i in clean.select_dtypes(include = 'object').columns:
    clean[i] = label.fit_transform(clean[i])
```

```python
clean
```

```python
clean.drop(['Customer ID', 'Purchase Date', 'Purchase Time'], axis = 1, inplace = True)
```

```python
clean
```

```python
clean['rating'] = clean['Customer Satisfaction'].apply(lambda x: 1 if x >= 8 else 0)
```

```python
clean
```

```python
x = clean.drop(['Customer Satisfaction', 'rating'], axis = 1)
y = clean['rating']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
```

```python
log = LogisticRegression().fit(X_train, y_train)
forest = RandomForestClassifier(n_estimators = 100, random_state = 42).fit(X_train, y_train)
svc = SVC().fit(X_train, y_train)
xgb = XGBClassifier().fit(X_train, y_train)
```

```python
y_pred1 = log.predict(X_test)
y_pred2 = forest.predict(X_test)
y_pred3 = svc.predict(X_test)
y_pred4 = xgb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, f1_score, recall_score, confusion_matrix
```

```python
models = [log, forest, svc, xgb]
preds = [y_pred1, y_pred2, y_pred3, y_pred4]
```

```python
scores = pd.Series(index = ['Logistic', 'Random Forest', 'SVC', 'XGB'], data = [accuracy_score(y_test, i) for i in preds])
```
