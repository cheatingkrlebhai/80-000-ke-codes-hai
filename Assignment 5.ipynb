```python
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X=df[["bachelors_degree"]]
Y=df["women_bachelors_degree"]
```

```python
model=LinearRegression()
```

```python
model.fit(X,Y)
```

```python
Y_pred=model.predict(X)
```

```python
A=mean_squared_error(Y,Y_pred)
```

```python
r2=r2_score(Y,Y_pred)
```

```python
print("MSE",A)
print("R2",r2)
```

```python
a=df[["high_school"]]
b=df["women_high_school"]
```

```python
model.fit(a,b)
```

```python
b_pred=model.predict(a)
```

```python
res=mean_squared_error(b,b_pred)
```

```python
R2=r2_score(b,b_pred)
```

```python
print("MSE:",res)
print("R2:",R2)
```

```python
#Insights:
#1. The wage premium for individuals with a bachelor's degree exhibits greater variability compared to those with a high school diploma.
#2. Men with a bachelor's degree tend to have a higher number of outliers in their wage premium when compared to men with a high school diploma.
#3. The scatterplot of wage premium for men with a bachelor's degree versus men with a high school diploma displays more dispersion than the scatterplot of wage premium for women with a bachelor's degree versus women with a high school diploma.
#4. Although women with a high school diploma earn less than women with a bachelor's degree, the former group exhibits fewer outliers in their wage premium compared to women with a bachelor's degree.
```
